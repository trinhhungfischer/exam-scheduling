```python
import pandas as pd
import numpy as np

# Read the data
df = pd.read_csv('aggregate.csv')

mean = df.groupby(['method', 'data', 'status'])['duration'].mean().reset_index()

mean['status'] = np.where((mean['method'] == 'cp') & (mean['status'] == 2), 1, mean['status'])
mean['status'] = np.where((mean['method'] == 'cp') & (mean['status'] == 4), 0, mean['status'])
mean.to_csv('mean_2.csv', index=False)
```

```python
sucess_solve = mean[mean['status'] == 0]
```

```python
def get_optimal_section_value(datapath: str, method:str):
    file_path = './' + method + "/set_conflict/" + datapath[:-4] + "_solution.csv"
    df = pd.read_csv(file_path)
    day = df.loc[df.index[-1], 'day']
    section = df.loc[df.index[-1], 'section']
    return day * 4 + section + 1

def get_optimal_day_value(datapath: str, method:str):
    file_path = './' + method + "/set_conflict/" + datapath[:-4] + "_solution.csv"
    df = pd.read_csv(file_path)
    day = df.loc[df.index[-1], 'day']
    section = df.loc[df.index[-1], 'section']
    return day + 1
```

```python
get_optimal_section_value('data_10_2_4_(0).csv', 'mip')
```

```python
sucess_solve['optimal_section'] = sucess_solve.apply(lambda x: get_optimal_section_value(x['data'], x['method']), axis=1)
sucess_solve['optimal_day'] = sucess_solve.apply(lambda x: get_optimal_day_value(x['data'], x['method']), axis=1)
```

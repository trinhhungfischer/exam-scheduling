```python
import numpy as np
mean = pd.read_csv('mean.csv')
mean['status'] = np.where((mean['method'] == 'cp') & (mean['status'] == 2), 1, mean['status'])
mean['status'] = np.where((mean['method'] == 'cp') & (mean['status'] == 4), 0, mean['status'])
```

```python
sucess_solve = mean[mean['status'] == 0]
```
